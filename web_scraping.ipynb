# Web Scraping

## 1. Create and send HTTP request to get HTML content

I use detik.com to get headline of news

```python
import requests
website_url = requests.get('https://www.detik.com/').text
```

## 2. use BeautifulSoup from python library to scraping

You can install it from pip

Then, if 'lxml' parser getting error. just install it.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('div',{'class': 'list-content'})
My_table
```
